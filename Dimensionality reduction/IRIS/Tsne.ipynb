```python
from sklearn import datasets
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import sklearn

from matplotlib import pyplot as plt
plt.rc('figure', max_open_warning = 0)
```

```python
iris = datasets.load_iris()
X = iris.data  
y = iris.target
X.shape,y.shape
```

```python
tsne=TSNE(n_components=2,perplexity =100,init='random',learning_rate='auto')
X_tsne=tsne.fit_transform(X)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()
ax.scatter(X_tsne[:,0], X_tsne[:,1],c=y,s=50)
```

```python
import math
from scipy.spatial import distance
from sklearn.metrics.pairwise import euclidean_distances
from sklearn import preprocessing
```

```python
def matrice_distance(data):
    d=euclidean_distances(data,data)
    d=preprocessing.normalize(d)
    return d
```

```python
A=matrice_distance(X_tsne)
B=d=matrice_distance(X)
```

```python
def estimateur(data,d,sigma):
    f=[]
    for i in range(40):
        s=0
        for j in range(40):
            s=s+math.exp(- (d[i][j]**2)*(sigma**(-1)))
        f.append(s)
    return f
```

```python
sigma=[0.01,0.1,1]

f=[estimateur(X,B,sigma) for sigma in sigma]

g=[estimateur(X_tsne,A,sigma) for sigma in sigma] 
```

```python
def KL(a, b):
    a = np.asarray(a, dtype=np.float64)
    b = np.asarray(b, dtype=np.float64)

    return np.sum(np.where(a != 0, a * np.log(a / b), 0))
```

```python
kl=[]
for i in range(3):
    kl.append(KL(f[i],g[i]))
```

```python
RMSE=[]

MSE = np.square(np.subtract(B,A)).mean()

RMSE.append(math.sqrt(MSE))
```

```python
from sklearn.manifold import trustworthiness

Trust=[]

Trust.append(sklearn.manifold.trustworthiness(B,A))
```

```python
Tab=pd.DataFrame([kl])
Tab.columns=['KL_0.01','KL_0.1','KL_1']
Tab['RMSE']=RMSE
Tab['Trust']=Trust
Tab
```
